```python
### test plot airsea data. Make sure the merged data looks reasonable. 
import numpy as np
import netCDF4 as nc
import matplotlib
import matplotlib.pyplot as plt
import datetime
import os
```

```python
### merged nc file info
year = '2022'
platf_num = '1031'
path_airsea = '/Users/chi/Documents/projects/sd-ni-wp/data_manipulate/data_merge_airsea/'
### load data
ds_airsea = nc.Dataset(path_airsea+'airsea-1min-merge-'+year+'-SD'+platf_num+'.nc')
time = ds_airsea.variables['time'][:]
print(ds_airsea.variables['time'].units)
dtime_airsea = np.array([datetime.datetime(int(year),1,1)+datetime.timedelta(seconds=time[i]) for i in range(len(time))])
uwnd = ds_airsea.variables['UWND_MEAN'][:]
vwnd = ds_airsea.variables['VWND_MEAN'][:]
```

```python
### merged nc file info
year = '2022'
platf_num = '1031'
path_airsea1 = '/Users/chi/Documents/projects/sd-ni-wp/data_manipulate/'
### load data
ds_airsea1 = nc.Dataset(path_airsea1+'airsea-raw-merge-'+year+'-SD'+platf_num+'.nc')
time = ds_airsea1.variables['time'][:]
print(ds_airsea1.variables['time'].units)
dtime_airsea1 = np.array([datetime.datetime(int(year),1,1)+datetime.timedelta(seconds=time[i]) for i in range(len(time))])
uwnd1 = ds_airsea1.variables['UWND_MEAN'][:]
vwnd1 = ds_airsea1.variables['VWND_MEAN'][:]
```

```python
tlim_plot = [datetime.datetime(int(year),7,15), datetime.datetime(int(year),11,1)]
plt.gcf().set_size_inches(18, 5)
# plt.plot(dtime_airsea, uwnd)
# plt.plot(dtime_airsea, vwnd)
plt.plot(dtime_airsea, 1.94384*np.sqrt( np.square(uwnd) + np.square(vwnd)) )
plt.plot(dtime_airsea1, 1.94384*np.sqrt( np.square(uwnd1) + np.square(vwnd1)) )
plt.xlim(tlim_plot)
plt.grid()
```
