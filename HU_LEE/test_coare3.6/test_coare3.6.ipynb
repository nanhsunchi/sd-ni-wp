```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import os
import sys
path_nhchi = os.path.expanduser('~/Documents/projects/analytical-tools/py_nhchi/')
# adding to the system path
sys.path.insert(0, path_nhchi)
from coare36vn_zrf_et import coare36vn_zrf_et
```

```python
u = np.arange(1,50) # mean wind speed accounting for the ocean current vector
zu = np.array([3.5])
test = np.ones( (len(u),14) )
print(u.shape,test[:,0].shape, zu.shape)
test_out = coare36vn_zrf_et(u,zu,27*test[:,0],zu,90*test[:,0],zu,1000*test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0])
print(test_out.shape)
nrow = 1; ncol = 2
plt.clf()
plt.rcParams.update({'font.size': 13})
fig, ax = plt.subplots(nrows=nrow,ncols=ncol)
fig.set_size_inches(13, 4)
''' plot u vs. u10'''
plt.subplot(nrow,ncol,1) 
plt.plot(u,label='U({:.1f}m)'.format(zu[0]))
plt.plot(test_out[:,32],label='U(10m)')
plt.grid(which='both',axis='both')
plt.ylabel('(m s$^{-1}$)')
plt.legend()
''' plot Cd '''
plt.subplot(nrow,ncol,2) 
plt.semilogy(test_out[:,34],label='Cdn10')
plt.grid(which='both',axis='both')
plt.xlabel('water-relative wind speed magnitudes (m s$^{-1}$) at 10m')
plt.ylabel('neautral value of Cd at 10m')
plt.legend()
''' settings '''
plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=0.3,hspace=0.3)
# plt.savefig('test_coare3.6(py)_u_cd.png',dpi=400,bbox_inches='tight')
```

```python
''' give a range of testing variable and see how it impacts the U10 & Cdn10 '''
var_str = 'rh'
var_test = np.linspace(30,100,50)
''' '''
u = np.arange(1,51) # mean wind speed accounting for the ocean current vector
zu = np.array([3.5])
test = np.ones( (len(u),14) )
print(u.shape,test[:,0].shape, zu.shape)
test_out = coare36vn_zrf_et(u,zu,27*test[:,0],zu,var_test,zu,1000*test[:,0],\
                            test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0],test[:,0])
''' plot '''
nrow = 1; ncol = 2
plt.clf()
plt.rcParams.update({'font.size': 13})
fig, ax = plt.subplots(nrows=nrow,ncols=ncol)
fig.set_size_inches(13, 4)
''' plot u vs. u10'''
plt.subplot(nrow,ncol,1) 
plt.plot(u,label='U({:.1f}m)'.format(zu[0]))
plt.plot(test_out[:,32],label='U(10m)')
plt.ylim([0,70])
plt.grid(which='both',axis='both')
plt.ylabel('(m s$^{-1}$)')
plt.legend()
''' plot Cd '''
plt.subplot(nrow,ncol,2) 
plt.semilogy(test_out[:,34],label='Cdn10')
plt.ylim([])
plt.grid(which='both',axis='both')
plt.xlabel('water-relative wind speed magnitudes (m s$^{-1}$) at 10m')
plt.ylabel('neautral value of Cd at 10m')
plt.legend()
''' settings '''
plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=0.3,hspace=0.3)
plt.suptitle('Varying '+var_str+' between {:.1f}~{:.1f}'.format(var_test[0],var_test[-1]))
```
