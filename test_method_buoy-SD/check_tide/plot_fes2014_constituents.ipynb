```python
import numpy as np
import netCDF4 as nc
import matplotlib
import matplotlib.pyplot as plt
import datetime
import glob
import os
```

```python
lonlim = [-90,-55]#[280,305]
latlim = [10,35]
pathU = os.path.expanduser('~/Documents/Data/AVISO/auxiliary/tide_model/fes2014/eastward_velocity/')
pathV = os.path.expanduser('~/Documents/Data/AVISO/auxiliary/tide_model/fes2014/northward_velocity/')
deglim = [-180,180]
vellim = [0,40] # cm/s
fileloc = glob.glob(pathU+'*.nc')
print(fileloc)
```

```python
### load topography data
path_topo = '/Users/chi/Documents/Data/Topography/GEBCO_24_Jul_2024/'
ds_topo = nc.Dataset(path_topo+'gebco_2024_n35.0_s5.0_w-92.0_e-50.0.nc')
# print(ds_topo)
lon_topo = ds_topo.variables['lon'][:]
lat_topo = ds_topo.variables['lat'][:]
elevation = ds_topo.variables['elevation'][:]
print(lon_topo.shape, lat_topo.shape, elevation.shape )
```

```python
varsnm = ['Ug','Ua','Vg','Va']
vars_str_title = ['U phaselag due to non equilibrium ocean tide at','U amplitude due to non equilibrium ocean tide at',\
                  'V phaselag due to non equilibrium ocean tide at','V amplitude due to non equilibrium ocean tide at']
cmap = ['hsv','turbo','hsv','turbo']
varlim = [[-180,180],[0,15],[-180,180],[0,15]]
x0_cb = [0.48, 0.905, 0.48, 0.905]
y0_cb = [0.56, 0.56, 0.14, 0.14]

for i in range(len(fileloc)):
    basename = os.path.basename(fileloc[i])
    du = nc.Dataset(pathU+basename)
    dv = nc.Dataset(pathV+basename)
    # print(dv)
    lat = du.variables['lat'][:]
    lon = du.variables['lon'][:]
    lon[lon> 180] = lon[lon> 180]-360
    Ug = du.variables['Ug'][:]
    Ua = du.variables['Ua'][:]
    Vg = dv.variables['Vg'][:]
    Va = dv.variables['Va'][:]

    ilon = (lon>= lonlim[0]) & (lon< lonlim[1])
    ilat = (lat>= latlim[0]) & (lat< latlim[1])
    ilon_topo = (lon_topo>= lonlim[0]) & (lon_topo< lonlim[1])
    ilat_topo = (lat_topo>= latlim[0]) & (lat_topo< latlim[1])
    ### plot
    plt.rcParams.update({'font.size': 14});
    plt.clf();
    fig, ax = plt.subplots(2,2);
    plt.gcf().set_size_inches(18, 16)
    for v in range( len(varsnm) ):
        plt.subplot(2,2,v+1)
        cs = plt.pcolormesh(lon[ilon],lat[ilat],eval(varsnm[v])[np.ix_(ilat,ilon)],cmap=cmap[v],vmin=varlim[v][0],vmax=varlim[v][1],zorder=1)
        plt.title(vars_str_title[v]+' '+basename[:-3]+' frequency',fontsize=14)
        plt.contour(lon_topo[ilon_topo], lat_topo[ilat_topo], elevation[np.ix_(ilat_topo,ilon_topo)], colors='k',levels=[0],zorder=2)
        plt.grid()
        ## add colorbar
        cbar_ax = fig.add_axes([x0_cb[v],y0_cb[v],.005,.3])
        axf = plt.colorbar(cs,orientation='vertical',cax=cbar_ax,extend='both')
    # break
    ### save figure
    fig.savefig('maps_AVISO_fes2014_UgUaVgVa_'+basename[:-3]+'_'+\
        str(lonlim[0])+'-'+str(lonlim[1])+'_'+str(latlim[0])+'-'+str(latlim[1])+'N.png',dpi=400,bbox_inches='tight')
```
